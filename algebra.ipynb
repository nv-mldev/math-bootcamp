```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
t = np.linspace(0,20,20)
p = 8000
r = 0.05
simple_interest = p*(1+r*t)
compound_interest = p*(1+r)**t
```

```python
plt.plot(t,simple_interest,'r', label='simple_interest')
plt.plot(t,compound_interest,'y', label='compound_interest')
plt.legend()
plt.show()
```
