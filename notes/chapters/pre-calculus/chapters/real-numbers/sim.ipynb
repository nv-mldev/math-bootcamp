```python
import numpy as np 
import matplotlib.pyplot as plt 
```

Ball bearings need to have extremely accurate sizes to work correctly. The ideal diameter of a particular ball bearing is $ 0.8\ $ cm, but a ball bearing is declared
acceptable if the error in the diameter size is less than $ 0.001 $ cm. Simulate the conditions

```python
np.random.seed(42)
diameters = np.random.normal(loc=0.8, scale=0.001, size=1000)
```

```python
plt.plot(diameters)
plt.xlabel('samples')
plt.ylabel('Diameter (cm)') 
plt.title("Samples vs Diameter")
plt.show()
```

```python
# Let's visualize the distribution of ball bearing diameters
plt.figure(figsize=(10, 6))

# Plot histogram with density plot
counts, bins, _ = plt.hist(diameters, bins=30, alpha=0.7, density=True, 
                           label='Simulated ball bearings')

# Add a line for the normal distribution curve
x = np.linspace(0.797, 0.803, 1000)
plt.plot(x, 1/(0.001 * np.sqrt(2 * np.pi)) * 
         np.exp(-(x - 0.8)**2 / (2 * 0.001**2)), 
         linewidth=2, color='red', label='Normal distribution curve')

# Add vertical lines for the acceptable range
plt.axvline(x=0.799, color='green', linestyle='--', 
            label='Acceptable range (±0.001 cm)')
plt.axvline(x=0.801, color='green', linestyle='--')

# Add vertical lines for the mean
plt.axvline(x=0.8, color='black', linestyle='-', label='Ideal diameter (0.8 cm)')

plt.title('Distribution of Ball Bearing Diameters')
plt.xlabel('Diameter (cm)')
plt.ylabel('Density')
plt.legend()
plt.grid(alpha=0.3)
plt.show()

# Calculate the percentage of ball bearings within acceptable range
acceptable = np.logical_and(diameters >= 0.799, diameters <= 0.801)
percentage_acceptable = np.mean(acceptable) * 100

print(f"Percentage of ball bearings within acceptable range: {percentage_acceptable:.2f}%")

# Calculate some basic statistics
mean_diameter = np.mean(diameters)
std_diameter = np.std(diameters)
min_diameter = np.min(diameters)
max_diameter = np.max(diameters)

print(f"Mean diameter: {mean_diameter:.6f} cm")
print(f"Standard deviation: {std_diameter:.6f} cm")
print(f"Minimum diameter: {min_diameter:.6f} cm")
print(f"Maximum diameter: {max_diameter:.6f} cm")
```

## Observations
1. **68-95-99.7 Rule**:
   - About 68% of values fall within 1 standard deviation of the mean
   - About 95% of values fall within 2 standard deviations
   - About 99.7% of values fall within 3 standard deviations

For your ball bearings with mean 0.8 and std dev 0.001:
- ~68% will have diameters between 0.799 and 0.801 cm
- ~95% will have diameters between 0.798 and 0.802 cm
- ~99.7% will have diameters between 0.797 and 0.803 cm
