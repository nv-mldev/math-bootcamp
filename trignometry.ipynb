```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
t = np.linspace(0,20,20)
p = 8000
r = 0.05
simple_interest = p*(1+r*t)
compound_interest = p*(1+r)**t
```

```python
plt.plot(t,simple_interest,'r', label='simple_interest')
plt.plot(t,compound_interest,'y', label='compound_interest')
plt.legend()
plt.show()
```

# Trignometric Functions
## plotting a sine and cosine wave between $ \left[-6\pi, 6\pi\right] $

```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python

x = np.linspace(-6*np.pi, 6*np.pi, 1000)
y = np.sin(x)
plt.figure(figsize=(12, 4))
plt.plot(x, y,'-r', label='sin(x)')
plt.grid(True)
plt.title('Sine Wave')
plt.xlabel(r'x (multiples of $\pi$)')
plt.ylabel(r'sin(x)')
ticks = np.arange(-6 * np.pi, 6.1 * np.pi, np.pi)  # Define the ticks array
plt.xticks(
    ticks=ticks,
    labels=[f"{int(t/np.pi)}$\\pi$" if t != 0 else "0" for t in ticks]
)
plt.show()
```

```python
x = np.linspace(-6*np.pi, 6*np.pi, 1000)
y = np.cos(x)
plt.figure(figsize=(12, 4))
plt.plot(x, y,'-g', label='cos(x)')
plt.grid(True)
plt.title('Cosine Wave')
plt.xlabel(r'x (multiples of $\pi$)')
plt.ylabel(r'cos(x)')
ticks = np.arange(-6 * np.pi, 6.1 * np.pi, np.pi)  # Define the ticks array
plt.xticks(
    ticks=ticks,
    labels=[f"{int(t/np.pi)}$\\pi$" if t != 0 else "0" for t in ticks]
)
plt.show()
```

```python
x = np.linspace(-5*np.pi/2, 5*np.pi/2, 1000)
y = np.tan(x)
plt.figure(figsize=(12, 4))
plt.plot(x, y,'-y', label='tan(x)')
plt.ylim(-10, 10)  # Limit y-axis to avoid extreme values
plt.grid(True)
plt.title('Tangent Wave')
plt.xlabel(r'x (multiples of $\pi$)')
plt.ylabel(r'tan(x)')
ticks = np.arange(-5 * np.pi/2, 5.3 * np.pi/2, np.pi/2)  # Define the ticks array
plt.xticks(
    ticks=ticks, labels=[f"{int(t/(np.pi/2))}$\\frac{{\\pi}}{{2}}$" if t != 0 else "0" for t in ticks])
plt.show()
```
