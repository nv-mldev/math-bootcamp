```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
x = np.arange(-4,5)
y = np.abs(x)
```

```python
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('$y = |x|$')
plt.show()
```

```python
x = np.linspace(-4,5,400)
y = 4*(5*x - x**2 - 2) /(x**2 + 2)
plt.plot(x,y)
plt.grid()
plt.title('$y = 4(5x - x^2 -2)/(x^2 +2)$')
plt.show()
```

```python
x = np.linspace(-1,1,100)
y = x**2 
plt.plot(x,y,label='$f(x) = x^2$')
plt.plot(x,y+0.5,'-r', label='$g(x) = f(x)+0.5$')
plt.plot(x,y-0.5,'-g', label='$g(x) = f(x)-0.5$')
plt.grid()
plt.legend()
plt.show()
```

```python
x = np.linspace(-1,1,100)
y = x**2 
plt.plot(x,y,label='$f(x) = x^2$')
plt.plot(x,2*y,'-r', label='$g(x) = 2x^2$')
plt.plot(x,0.5*y,'-g', label='$h(x) = 0.5x^2$')
plt.grid()
plt.legend()
plt.show()
```

```python
x = np.linspace(-1,1,100)
y = x**2 
plt.plot(x,y,label='$f(x) = x^2$')
plt.plot(x,-y,'-r', label='$g(x) = -x^2$')
plt.grid()
plt.legend()
plt.show()
```

```python
x = np.linspace(-2,2,100)
y = x**2 
plt.plot(x,y,label='$f(x) = x^2$')
plt.plot(x,(x+1)**2,'-r', label='$g(x) = (x+1)^2$')
plt.plot(x,(x-1)**2,'-g', label='$h(x) = (x-1)^2$')
plt.xlim([-2,2])
plt.grid()
plt.legend()
plt.show()
```
