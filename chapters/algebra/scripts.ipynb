```python
import numpy as np
import matplotlib.pyplot as plt

# Define the x range
x = np.linspace(-3, 3, 400)

# Define polynomials with coefficients (highest degree first):
# Degree 0: f(x) = 5
# Degree 1: f(x) = 2x + 1
# Degree 2: f(x) = x^2 - 3x + 2
# Degree 4: f(x) = x^4 - 5x^2 + 4
# Degree 7: f(x) = x^7 - 3x^5 + 2x^3 - x + 1
polynomials = {
    0: [5],
    1: [2, 1],
    2: [1, -3, 2],
    4: [1, -1, -7, 6, 4],
    7: [1, 0, -14, 0, 49, 0, -36,0]
}

# Create a figure with one subplot per polynomial
fig, axs = plt.subplots(len(polynomials), 1, figsize=(8, 20))
fig.suptitle("Graphs of Polynomials with Various Degrees", fontsize=16)

for ax, (degree, coeffs) in zip(axs, polynomials.items()):
    # Evaluate the polynomial using numpy.polyval
    y = np.polyval(coeffs, x)
    
    # Build a string representation of the polynomial for the legend
    poly_str = ""
    for i, coef in enumerate(coeffs):
        exp = degree - i
        # Skip zero coefficients
        if coef == 0:
            continue
        # Add a plus sign for positive coefficients (if not the first term)
        sign = " + " if coef > 0 and i > 0 else ""
        if exp > 1:
            poly_str += f"{sign}{coef}x^{exp}"
        elif exp == 1:
            poly_str += f"{sign}{coef}x"
        else:
            poly_str += f"{sign}{coef}"
    
    # Plot the polynomial
    ax.plot(x, y, label=f"f(x) = {poly_str}")
    ax.set_title(f"Degree {degree} Polynomial")
    ax.set_xlabel("x")
    ax.set_ylabel("f(x)")
    ax.legend()
    ax.grid(True)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the x range
x = np.linspace(-3, 3, 400)

# Define polynomials with coefficients (highest degree first):
# Degree 0: f(x) = 5
# Degree 1: f(x) = 2x + 1
# Degree 2: f(x) = x^2 - 3x + 2
# Degree 4: f(x) = x^4 - 5x^2 + 4
# Degree 7: f(x) = x^7 - 3x^5 + 2x^3 - x + 1
polynomials = {
    0: [5],
    1: [2, 1],
    2: [1, -3, 2],
    4: [1, -1, -7, 6, 4],
    7: [1, 0, -14, 0, 49, 0, -36, 0]
}

for degree, coeffs in polynomials.items():
    # Evaluate the polynomial using numpy.polyval
    y = np.polyval(coeffs, x)
    
    # Build a string representation of the polynomial for the legend
    poly_str = ""
    for i, coef in enumerate(coeffs):
        exp = degree - i
        # Skip zero coefficients
        if coef == 0:
            continue
        # Add a plus sign for positive coefficients (if not the first term)
        sign = " + " if coef > 0 and i > 0 else ""
        if exp > 1:
            poly_str += f"{sign}{coef}x^{exp}"
        elif exp == 1:
            poly_str += f"{sign}{coef}x"
        else:
            poly_str += f"{sign}{coef}"
    
    # Create a new figure for each polynomial
    plt.figure()
    plt.plot(x, y, label=f"f(x) = {poly_str}")
    plt.title(f"Degree {degree} Polynomial")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    
    # Save the figure as a PNG file
    filename = f"polynomial_degree_{degree}.png"
    plt.savefig(filename)
    plt.close()  # Close the figure after saving to free up memory

print("Figures saved successfully!")
```
